```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('helium_fn.csv') #628s
data = df.to_numpy()
E_exact = -2.175224377034119598311159245194404
w = 200
E = data[:,0]
E_av = np.mean(E[w:])
E_err = np.std(E[w:])/np.sqrt(len(E[w:]))
```

```python
print('Final energy',E_av,'pm',E_err)
```
