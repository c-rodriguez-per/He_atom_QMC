```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('helium.csv')
data = df.to_numpy()
E_exact = -2.175224377034119598311159245194404
min_index = np.argmin(data[:,7])
#Let us extract our columns
alpha = data[:,0]
beta = data[:,1]
w_ep = data[:,2]
w_ee = data[:,3]
f_ep = data[:,4]
f_ee = data[:,5]
acceptance = data[:,6]
E = data[:,7]
E_errbar = data[:,8]
w_bf = data[:,9]
f_bf = data[:,10]

#Parameters
alpha_min = alpha[min_index]
beta_min = beta[min_index]
w_ep_min = w_ep[min_index]
w_ee_min = w_ee[min_index]
```

```python
print('Emin at:',E[min_index], 'Var:', E_errbar[min_index])
print('With parameters, a=',alpha[min_index],', b=',beta[min_index],', wep=',w_ep[min_index], 'wee=', w_ee_min, ', fep=',f_ep[min_index], 'fee=', f_ee[min_index],
      'wbf=', w_bf[min_index], ', fbf=',f_bf[min_index],)
print('Acceptance:', acceptance[min_index])
```

```python
#BF optimization
plt.subplot(1,2,1)
Ebf = E[f_bf == f_bf[min_index]]
wbf = w_bf[f_bf == f_bf[min_index]]
plt.scatter(wbf,Ebf)
plt.title('WBF')
plt.subplot(1,2,2)
Ebf = E[w_bf == w_bf[min_index]]
fbf = f_bf[w_bf == w_bf[min_index]]
plt.scatter(fbf,Ebf)
plt.title('FBF')
```
