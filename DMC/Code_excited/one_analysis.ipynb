```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#df = pd.read_csv('excited_dmc_one.csv') #628s
df = pd.read_csv('dmc_big.csv')
data = df.to_numpy()
E_exact = -2.175224377034119598311159245194404
w = 1
step = data[:,1]
E = data[:,2]#[data[:,0]==0.001]
E_av = np.mean(E[w:])
E_err = np.std(E[w:])/np.sqrt(len(E[w:]))
pop = data[:,5]
tau = data[:, 0]
step = data[:,1]
```
